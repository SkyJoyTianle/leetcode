```python
a = 20
b = 20
def comb(n):
  return (n*(n-1))//2
def cal(a,b):
  if a <= 0 or b < 0:
    return 0
  if b == 0:
    return 1
  total = comb(a + b)
  p1 = a * b / total      #1蓝1红概率
  p2 = comb(a) / total    #2红概率
  p3 = comb(b) / total    #2蓝概率
  #x = p1*cal(a,b-1)+p2*cal(a-2,b)+p3*x
  return (p1 * cal(a, b - 1) + p2 * cal(a-2, b))/(1 - p3)

cal(a, b)
```

```python
def fr(n):
  return (n*(n-1))//2

m, n = 50, 50

a = [[0] * (n + 1) for i in range(m + 1)]

for i in range(1, m + 1):
  a[i][0] = 1

for j in range(1, n + 1):
  a[0][j] = 0

for i in range(1, m + 1):
  for j in range(1, n + 1):
    total = fr(i + j)
    p1 = i * j / total
    p2 = fr(i) / total
    p3 = fr(j) / total
    if i >= 2:
      w = p1 * a[i][j - 1] + p2 * a[i - 2][j]
    else:
      w = p1 * a[i][j - 1]
    a[i][j] = w / (1 - p3)

print(a[-1][-1])
```
