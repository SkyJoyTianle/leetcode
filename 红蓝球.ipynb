```python
a = 20
b = 20
def comb(n):
  return (n*(n-1))//2
def cal(a,b):
  if a <= 0 or b < 0:
    return 0
  if b == 0:
    return 1
  total = comb(a + b)
  p1 = a * b / total      #1蓝1红概率
  p2 = comb(a) / total    #2红概率
  p3 = comb(b) / total    #2蓝概率
  #x = p1*cal(a,b-1)+p2*cal(a-2,b)+p3*x
  return (p1 * cal(a, b - 1) + p2 * cal(a-2, b))/(1 - p3)

cal(a, b)
```
